```python
# UNO only with only colors
# Give 5 cards to two players
# the deck consist of 32 cards, which 1/4 of the deck is 'r' = red 'g' = green , 'b' = blue, 'y' = yellow
# the bottom card is flipped and the game start
# if the player has a card with the color, the the player put the card
# if no card can be play
# the winner is the player with the less cards

import random

class Player:
    def __init__(self):
        self.my_cards = []
        self.card = None
        self.turn = None
        self.card_index = None
        self.paso = False
    
    def give_me_cards(self, cards):
        """Get players cards"""
        self.my_cards = cards
    
    def mano(self):
        """return the len of the hand"""
        return len(self.my_cards)
    
    def show_cards(self):
        """print the player cards"""
        for x in self.my_cards:
            print(x,end=",")
        print()
        
    def play(self, visual):
        """Play the game, the playe can see the 'visual' card"""
        print("VISUAL=",visual)
        if visual in self.my_cards:                # I have a card can play
            print("YES I have")
            self.card_index = self.my_cards.index(visual)     # get index of card
            self.card = self.my_cards.pop(self.card_index)
            return self.card
        else:
            print("NO I dont have")
            self.paso = True
            return visual # return the card that is visual / since paso / no action
        return None
            
    
class Deck:
    def __init__(self):
        self.cards = []
        self.colors = ['r', 'g', 'b', 'y']
    
    def create_deck(self):
        """Create a deck and shuffle it"""
        for i in range(8):
            for x in self.colors:
                self.cards.append(x)
        random.shuffle(self.cards)
        return self.cards
    
    
class Uno():
    def __init__(self):
        self.players = 2
        self.player1 = Player()
        self.player2 = Player()
        
        self.cards = []
        self.deck = Deck()
        self.cards = self.deck.create_deck()
        print("=================")
        print(self.cards)
        print("=================")

    def reparte(self):
        """Give the cards to the players"""
        self.player1.give_me_cards((self.cards[:5]))
        self.player2.give_me_cards((self.cards[5:10]))
        self.cards = self.cards[10:]
        print("-----------------")
        print(self.cards)
        print("-----------------")

    def play_uno(self):
        """Play simplified UNO"""
        self.reparte()            # give to the players 7 cards
        start_card = self.cards.pop()#self.cards[-1]    # the last card is the start
        
        self.player1.show_cards()
        self.player2.show_cards()
        
        print("==>",start_card)
        print(self.cards)
        print("*****************")
        
        while True:# Play until both player pass and check who is the winner

            start_card = self.player1.play(start_card)
            self.player1.show_cards()
            print("Player1 paso:", self.player1.paso)
            
            start_card = self.player2.play(start_card)
            self.player2.show_cards()
            print("Player2 paso:", self.player2.paso)
            
            if self.player1.paso == True and self.player2.paso == True:
                break
        
        if self.player1.mano() < self.player2.mano():
            return 1
        if self.player1.mano() > self.player2.mano():
            return 2
        if self.player1.mano() == self.player2.mano():
            return 0
        
    
deck = Deck()
deck.create_deck()
uno = Uno()
winner = uno.play_uno()
print()
if winner == 1:
    print("Player 1 is the winner")
elif winner == 2:
    print("Player 2 is the winner")
else:
    print("NONE won")
```
